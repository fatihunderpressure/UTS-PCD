```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# baca gambar
img = cv2.imread('/content/cool.png')

if img is None:
    print("file ga ketemu goblok 💀")
else:
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # kernel
    lpf = (1/9) * np.array([[1,1,1],[1,1,1],[1,1,1]])
    hpf = np.array([[-1,-1,-1],[-1,8,-1],[-1,-1,-1]])
    bpf = np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])

    # konvolusi
    img_lpf = cv2.filter2D(img_gray, -1, lpf)
    img_hpf = cv2.filter2D(img_gray, -1, hpf)
    img_bpf = cv2.filter2D(img_gray, -1, bpf)

    # tampil
    titles = ['Original','LPF','HPF','BPF']
    images = [img_rgb, img_lpf, img_hpf, img_bpf]

    for i in range(4):
        plt.subplot(2,2,i+1)
        if i == 0:
            plt.imshow(images[i])
        else:
            plt.imshow(images[i], cmap='gray')
        plt.title(titles[i])
        plt.axis('off')

    plt.show()
```
