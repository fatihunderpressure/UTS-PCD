```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('trio maut.png')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

img_float = img_gray.astype(np.float32)

# negatif
neg = 255 - img_gray

# log
c = 255 / np.log(1 + np.max(img_float))
log_img = c * np.log(1 + img_float)
log_img = cv2.normalize(log_img, None, 0, 255, cv2.NORM_MINMAX)
log_img = np.uint8(log_img)

# tampil 4 gambar
titles = ['Asli','Grayscale','Negatif','Log']
images = [cv2.cvtColor(img, cv2.COLOR_BGR2RGB), img_gray, neg, log_img]

for i in range(4):
    plt.subplot(2,4,i+1)
    if i == 0:
        plt.imshow(images[i])
    else:
        plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')

# histogram
for i in range(4):
    plt.subplot(2,4,i+5)
    if i == 0:
        plt.hist(images[i].ravel(),256)
    else:
        plt.hist(images[i].ravel(),256,[0,256])
    plt.title('Hist ' + titles[i])

plt.show()
```
